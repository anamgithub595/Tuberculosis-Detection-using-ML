```python
def split_data(SOURCE, TRAINING, VAL, TESTING, SPLIT_SIZE_TRAIN = 0.8, SPLIT_SIZE_VAL = 0.1):

   
    
    use_dir = [fn for fn in os.listdir(SOURCE)] 
    
    
    random.seed(2)
    
    train_dt = random.sample(use_dir, np.int64(SPLIT_SIZE_TRAIN * len(use_dir)))
    
    test_dt = [fn for fn in use_dir if fn not in train_dt]
    
    val_dt = random.sample(test_dt, np.int64(SPLIT_SIZE_VAL * len(use_dir)))
    
    test_dt = [fn for fn in test_dt if fn not in val_dt]
    
    for fn in train_dt :
        shutil.copy(os.path.join(SOURCE, fn), TRAINING)
    
    for fn in val_dt :
        shutil.copy(os.path.join(SOURCE, fn), VAL)
    
    for fn in test_dt:
        shutil.copy(os.path.join(SOURCE, fn), TESTING)
```

```python
 def copy_Xray_images(SOURCE_Normal, SOURCE_TB, D_IMAGE_DIR):
        
        
    Source_dir_n = [fn for fn in os.listdir(SOURCE_Normal)]
    
    Source_dir_tb = [fn for fn in os.listdir(SOURCE_TB)]
    
    random.seed(2)
    
    for fn in Source_dir_n:
        shutil.copy(os.path.join(SOURCE_Normal, fn), D_IMAGE_DIR)
    
    for fn in Source_dir_tb :
        shutil.copy(os.path.join(SOURCE_TB, fn), D_IMAGE_DIR)
```

```python
import matplotlib.pyplot as plt
from tensorflow import keras
from keras.models import *
from keras.layers import *
from keras.preprocessing import image
from keras import optimizers
```

```python
#CNN based model

model = Sequential()
model.add(Conv2D(34, kernel_size = (3,3), activation = 'relu', input_shape = (224,224,3)))
model.add(Conv2D(64,(3,3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(0.25))


model.add(Conv2D(64,(3,3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(0.25))


model.add(Conv2D(128,(3,3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(0.25))


model.add(Flatten())
model.add(Dense(64, activation = 'relu'))
model.add(Dropout(0.25))
model.add(Dense(1, activation = 'sigmoid'))


model.compile(loss = keras.losses.binary_crossentropy, optimizer = 'adam', metrics = ['accuracy'])

```

```python
#Training

train_datagen = image.ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
)

test_datagen = image.ImageDataGenerator(
    rescale=1./255,
)
```

```python
train_generator = train_datagen.flow_from_directory(
    '/Users/apple/Downloads/DataSet_TB/training',
    target_size = (224,224),
    batch_size = 32,
    class_mode = 'binary'
)
```

```python
validation_generator = test_datagen.flow_from_directory(
    '/Users/apple/Downloads/DataSet_TB/validation',
    target_size = (224,224),
    batch_size = 16,
    class_mode = 'binary'
)
```

```python
test_generator = test_datagen.flow_from_directory(
    '/Users/apple/Downloads/DataSet_TB/testing',
    target_size = (224,224),
    batch_size = 16,
    class_mode = 'binary'
)
```

```python
hist = model.fit_generator(
  
    train_generator,
    steps_per_epoch = 5,
    epochs = 4,
    validation_data = validation_generator,
    validation_steps = 2, 
)
```
